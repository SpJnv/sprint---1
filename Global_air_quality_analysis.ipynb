#### **Dataset Description:**
The dataset, titled "global_air_quality.csv," consists of 10,000 records covering air quality measurements from different cities around the world. The dataset includes the following columns:

1. **City:** The name of the city where the air quality measurements were taken.
2. **Country:** The country of the city.
3. **Date:** The date on which the measurements were recorded.
4. **PM2.5:** Particulate matter with a diameter of less than 2.5 micrometers. PM2.5 can penetrate the respiratory system, causing serious health issues such as heart and lung diseases. It also contributes to environmental problems like haze.
5. **PM10:** Particulate matter with a diameter of less than 10 micrometers.  PM10 particles can also cause respiratory issues and reduce visibility, contributing to pollution-related problems.
6. **NO2:** Nitrogen dioxide concentration. A toxic gas produced from vehicle emissions, industrial processes, and other combustion sources. High levels of NO2 can cause respiratory problems, contribute to the formation of ground-level ozone, and negatively impact vegetation.
7. **SO2:** Sulfur dioxide concentration. A gas produced by burning fossil fuels containing sulfur, such as coal or oil. SO2 can cause acid rain, which harms ecosystems, buildings, and human health.
8. **CO:** Carbon monoxide concentration. A colorless, odorless gas produced by burning fossil fuels. High levels of CO can cause harmful effects on human health, including impaired cognitive function and cardiovascular issues.
9. **O3:** Ozone concentration. Ground-level ozone is a harmful pollutant formed when NO2 and volatile organic compounds react in sunlight. High levels of ozone can lead to respiratory issues, aggravate asthma, and damage crops and other vegetation.
10. **Temperature:** Ambient temperature during the measurement.
11. **Humidity:** Relative humidity during the measurement.
12. **Wind Speed:** Wind speed during the measurement.

This dataset provides a comprehensive view of various air pollutants and weather conditions across different geographical locations, allowing for in-depth analysis and visualization.

### **Data Visualization**

#### 1. Distribution of Pollutants
Visualization:
We'll start by plotting histograms for key pollutants like PM2.5, PM10, NO2, SO2, CO, and O3 to understand their distributions.

```python
import matplotlib.pyplot as plt

# List of pollutants to visualize
pollutants = ['PM2.5', 'PM10', 'NO2', 'SO2', 'CO', 'O3']

# Create histograms for each pollutant
df[pollutants].hist(bins=30, figsize=(12, 10), edgecolor="b")
plt.suptitle('Distribution of Pollutants' "\n"'Each x axis shows range of dataset values and y axis shows frequency')
plt.show()
```

##### Insights:
**`PM2.5 Distribution:`** The histogram shows that the majority of PM2.5 values are clustered around the lower end, with a significant number of readings between 0 to 50 µg/m³. However, there are some extreme values above 100 µg/m³, indicating periods or areas of severe pollution.

**`NO2 Distribution:`** The NO2 histogram reveals that most measurements are below 50 µg/m³, but there are a few peaks at higher levels, suggesting occasional spikes in NO2 concentrations.

**`O3 Distribution:`** The O3 histogram is more spread out, with concentrations commonly ranging between 10 to 100 µg/m³. This indicates variability in ozone levels, possibly due to different environmental conditions.

#### 2. Air Quality by Country
Visualization:
Next, we'll plot the average PM2.5 levels by country using a bar plot. This will help us compare air quality across different countries.

```python
# Group by country and calculate the mean for PM2.5
avg_pm25_by_country = df.groupby('Country')['PM2.5'].mean().sort_values()

# Plotting
plt.figure(figsize=(10, 8))
avg_pm25_by_country.plot(kind='barh', color='skyblue')
plt.title('Average PM2.5 Levels by Country', fontsize=16)
plt.xlabel('Average PM2.5 Concentration')
plt.ylabel('Country')
plt.grid(axis = "x", color = "green", linestyle = "--", alpha = 0.30)
plt.show()
```

##### Insights:
**`Top Polluted Country:`** The bar plot reveals that Country `UAE` has the highest average PM2.5 level, at around 85 µg/m³. This indicates poor air quality and potential health risks in that country.

**`Cleanest Air:`** Country `France` has the lowest average PM2.5 level, at around  75 µg/m³. This suggests that Country Y has effective air quality management and fewer sources of pollution.

**`Mid-Range Countries:`** Several countries like china, russia, canada fall in the middle range, with PM2.5 levels between 75 to 80 µg/m³. These countries may experience moderate air pollution, potentially from a mix of urban and industrial sources.

### Compare the average temperature, humidity, and wind speed across the 19 unique countries.

```python
data = pd.read_csv(r"C:\Users\shiva\Downloads\global_air_quality.csv")
# Convert necessary columns to numeric to ensure no errors during aggregation
data['Temperature'] = pd.to_numeric(data['Temperature'], errors='coerce')
data['Humidity'] = pd.to_numeric(data['Humidity'], errors='coerce')
data['Wind Speed'] = pd.to_numeric(data['Wind Speed'], errors='coerce')

# Group the data by 'Country' and calculate the mean for each
grouped_data = data.groupby('Country').mean(numeric_only=True).reset_index()

# Extract the grouped data
countries = grouped_data['Country']
temperature = grouped_data['Temperature']
humidity = grouped_data['Humidity']
wind_speed = grouped_data['Wind Speed']

# Set up the bar width and positions
bar_width = 0.25
x_indexes = np.arange(len(countries))

# Plotting the Grouped Bar Chart
plt.figure(figsize=(15, 8))

plt.bar(x_indexes, temperature, width=bar_width, color='r', label='Temperature')
plt.bar(x_indexes + bar_width, humidity, width=bar_width, color='b', label='Humidity')
plt.bar(x_indexes + 2 * bar_width, wind_speed, width=bar_width, color='g', label='Wind Speed')

# Adding labels and title
plt.xlabel('Countries', fontsize=14)
plt.ylabel('Values', fontsize=14)
plt.title('Average Temperature, Humidity, and Wind Speed across Unique Countries', fontsize=16)
plt.xticks(ticks=x_indexes + bar_width, labels=countries, rotation=90, fontsize=12)

plt.legend()
plt.tight_layout()
plt.show()
```

##### Insights
1. Countries like Brazil and India have higher average temperatures, while countries like Canada and Australia have more moderate temperatures.
2. Humidity and wind speed show less variation compared to temperature.

#### Visualize the relationship between temperature and humidity for each country.

```python
# Load the dataset
data = pd.read_csv(r"C:\Users\shiva\Downloads\global_air_quality.csv")

# Convert necessary columns to numeric
data['Temperature'] = pd.to_numeric(data['Temperature'], errors='coerce')
data['Humidity'] = pd.to_numeric(data['Humidity'], errors='coerce')

# Drop rows with NaN values in these columns
cleaned_data = data.dropna(subset=['Temperature', 'Humidity'])

# Group the data by 'Country' and calculate the mean for each
grouped_data = cleaned_data.groupby('Country').mean(numeric_only=True).reset_index()

# Scatter Plot: Temperature vs Humidity
plt.figure(figsize=(10, 6))
plt.scatter(grouped_data['Temperature'], grouped_data['Humidity'], color='purple', s=100)

# Adding labels and title
plt.xlabel('Temperature (°C)', fontsize=14)
plt.ylabel('Humidity (%)', fontsize=14)
plt.title('Temperature vs Humidity across Unique Countries', fontsize=16)

for i, country in enumerate(grouped_data['Country']):
    plt.text(grouped_data['Temperature'][i], grouped_data['Humidity'][i], country, fontsize=9)

plt.grid(True)
plt.tight_layout()
plt.show()
```

#### Insights
1. Most countries are clustered around moderate humidity levels, with varying temperatures.
2. Brazil stands out with higher temperature and moderate humidity.
3. And Thailand stands out with highest humidity and moderate temperature

#### Show the distribution of wind speed across the different countries

```python
# Load the dataset
data = pd.read_csv(r"C:\Users\shiva\Downloads\global_air_quality.csv")

# Convert necessary columns to numeric
data['Wind Speed'] = pd.to_numeric(data['Wind Speed'], errors='coerce')

# Group the data by 'Country' and calculate the mean for each
grouped_data = data.groupby('Country').mean(numeric_only=True).reset_index()

# Box Plot: Distribution of Wind Speed across Countries
plt.figure(figsize=(12, 6))
plt.boxplot([grouped_data['Wind Speed']], vert=False, patch_artist=True, 
            boxprops=dict(facecolor='lightblue', color='blue'),
            medianprops=dict(color='red'))

# Adding labels and title
plt.xlabel('Wind Speed (m/s)', fontsize=14)
plt.yticks([1], ['Countries'])
plt.title('Distribution of Wind Speed across Unique Countries', fontsize=16)

plt.tight_layout()
plt.show()
```

#### Insights
1. The box plot displays the distribution of wind speed across all the countries.
2. The spread of wind speeds is relatively consistent across countries, with a few outliers.
3. The median wind speed is around 10 m/s, with a range extending from lower to higher wind speeds, indicating variability in wind conditions among the countries.
